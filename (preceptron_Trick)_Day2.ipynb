# üß† Perceptron Algorithm ‚Äì Step-by-Step Guide

The **Perceptron** is the most basic unit of a neural network. It is used for **binary classification** (e.g., yes/no, spam/not spam).

---

## üîç What It Does:
It takes numerical input features, applies weights and bias, calculates a weighted sum, and uses an activation function to give an output of `0` or `1`.

---

## ‚úÖ Step-by-Step Process

### 1. **Input Features**
Let:  
$x_1, x_2, ..., x_n$ be the input values.  
Example: Study hours, Sleep hours

### 2. **Weights**
Assign each input a weight:  
$w_1, w_2, ..., w_n$  
These represent how important each feature is.

### 3. **Bias**
Add a bias term $b$ to shift the decision boundary.

### 4. **Calculate Weighted Sum**

$$
z = (w_1 \cdot x_1) + (w_2 \cdot x_2) + ... + (w_n \cdot x_n) + b
$$

### 5. **Activation Function (Step Function)**

$$
\text{output} =
\begin{cases}
1 & \text{if } z \geq 0 \\
0 & \text{if } z < 0
\end{cases}
$$

---

## üéØ Example

Inputs:  
$x_1 = 4$ (study hours), $x_2 = 6$ (sleep hours)  
Weights:  
$w_1 = 0.6$, $w_2 = 0.3$  
Bias:  
$b = -3$

**Weighted sum:**

$$
z = (0.6 \cdot 4) + (0.3 \cdot 6) - 3 = 2.4 + 1.8 - 3 = 1.2
$$

Since $z = 1.2 \geq 0$ ‚Üí Output = **1 (Pass)** ‚úÖ

---

## üìå Summary

- Multiply inputs by weights  
- Add bias  
- Apply activation function  
- Output 0 or 1  
- Simple but powerful in linearly separable cases!


```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
